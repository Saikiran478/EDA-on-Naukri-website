```python
avg_sal = df.groupby(["Job_name"],as_index =False)["Minimum Salary in LPA"].mean().sort_values(by = "Minimum Salary in LPA", ascending = False)
a = sns.barplot(data = avg_sal,x  = avg_sal["Job_name"], y = avg_sal["Minimum Salary in LPA"])
plt.xlabel("Job Role")
plt.ylabel("Average Salary for Job Type")
plt.xticks(rotation = 90)
for i in a.containers:
    a.bar_label(i)
plt.show()
```

#### The Data Science are getting higher salary and web and software developer are getting lower salry package according to this data

```python
plt.figure(figsize = (8,6))
avg_rating = df.groupby(["Job_name"],as_index =False)["Rating"].mean().sort_values(by ="Rating")
a1 = sns.barplot(data = avg_rating,x  = avg_rating["Job_name"], y = avg_rating["Rating"])
plt.xlabel("Job Role")
plt.ylabel("Average Rating for Job Type")
plt.xticks(rotation = 90)
for i in a1.containers:
    a1.bar_label(i)
plt.show()
```

#### Software developer, Data science jobs are having more rating and web developer and full stack developer is having less rating
