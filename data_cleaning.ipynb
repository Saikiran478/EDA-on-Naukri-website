```python
a = df1.groupby("Job_name")
a.first()
```

```python
d_sd = a.get_group("software-developer")
d_sd
```

```python
sd = d_sd[d_sd.isnull().sum(axis = 1) == 0]
sd                                           # df of group software developer
```

```python
mean_sd = sd["Rating"].mean()
mean_sd                             # mean of rating in software developer df
```

```python
sd1 = d_sd[d_sd.isnull().sum(axis = 1) == 1]
sd1                                            # DF which are having nan values in rating column 
```

```python
sd_1 = sd1.head(10)                             # 10 rows from nan valued DF
```

```python
add = (sd,sd_1)
r = pd.concat(add)
df_sd = r.sort_index()                          # adding 10 rows from nan valued DF and non null DF 
```

```python
df_sd.info()
```

```python
df_sd['Rating'].fillna(value=mean_sd, inplace=True)                           # fill the 10 rows with mean of rating
```

```python
df_sd["Job_name"] = "Software Developer"
df_sd
```
